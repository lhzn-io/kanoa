```python
import kanoa
from kanoa.utils import log_debug, log_info, log_stream, log_warning

# Enable verbose logging
kanoa.options.verbose = 2
```

`log_info()` invocations write messages into a shaded box, one per notebook cell execution

the background color can be updated with `kanoa.options.user_log_bg_color`

```python
log_info(
    f"First message, default background color is {kanoa.options.user_log_bg_color}"
)
log_info(
    "Second message, they are always *plain* text like _print_ statements, not markdown"
)
log_warning("Third message with a bolded title", title="Important Notice")
```

Named containers can be created using `log_stream()`.

User streams default to `user_log_bg_color` (gray), while kanoa internals use `internal_log_bg_color` (lavender).

```python
log_info("This message is outside any log stream")

# new cell generates a new container
with log_stream():
    log_info("Fourth message, in a log stream")
    log_info("Fifth message")
```

Alternating with blocks result in new containers, we deferred state-tracking and other advanced features.

Maybe we revisit the idea in a separate `kanoa-logging` module some day if there's a lot of interest

```python
with log_stream():
    log_info("Sixth message")

with log_stream(title="Processing Data Pipeline"):
    log_info("Step 1: Loading data from source...")
    log_debug("Found 1,234 records", title="Debug Info")
    log_warning("Step 2: Detected 5 missing values", title="Data Quality")
    log_info("Step 3: Applying transformations...")
    log_info("Step 4: Complete! 1,234 records processed.")

with log_stream():
    log_info("Seventh message")

with log_stream(title="Processing Data Pipeline"):
    log_info("Step 5: World domination...")
```

## Test 3: Custom Colors

You can override the background color:

```python
# Ocean blue with custom opacity
with log_stream(title="Ocean Theme", bg_color=(2, 62, 138), bg_opacity=0.2):
    log_info("This uses ocean blue colors")
    log_info("With 20% background opacity")

# Sunset orange with higher opacity
with log_stream(title="Sunset Theme", bg_color=(230, 115, 0), bg_opacity=0.15):
    log_info("This uses sunset orange colors")
    log_info("With 15% background opacity")
```

## Test 4: Real Interpretation Example

Test with actual analytics interpretation:

```python
import matplotlib.pyplot as plt
import numpy as np

from kanoa import AnalyticsInterpreter

# Create sample data
x = np.linspace(0, 10, 100)
y = np.sin(x) + 0.1 * np.random.randn(100)

# Create figure
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, "o-", alpha=0.7)
ax.set_title("Noisy Sine Wave")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nNOTE: The interpretation below should show all log messages")
print("in a SINGLE lavender box, followed by the AI result.\n")
```

```python
# This should now show all logs in one unified box
interpreter = AnalyticsInterpreter(
    backend="gemini", model="gemini-2.5-flash", verbose=True
)

result = interpreter.interpret(
    fig=fig,
    context="Noisy sine wave data",
    focus="Describe the pattern and noise characteristics",
    display_result=True,
)
```
